```python
import cv2

# Open the webcam
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # If the frame was captured successfully
    if not ret:
        print("Error: Could not read frame.")
        break

    # Display the frame
    cv2.imshow("Webcam Feed", frame)

    # Press 'q' to exit the loop and close the window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 as cv
import os

# Path to the folder containing images
folder_path = 'photos/'  # Replace with your folder path

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    # Full path to the image
    img_path = os.path.join(folder_path, filename)
    
    # Check if the file is an image (you can add more extensions if needed)
    if filename.endswith(('.jpg', '.jpeg', '.png')):
        # Read and display the image
        img = cv.imread(img_path)
        
        # Check if the image was loaded successfully
        if img is None:
            print(f"Failed to load image: {filename}")
            continue
        
        cv.imshow('Image', img)
        
        # Wait for a key press to display the next image
        # You can adjust the wait time (in milliseconds) for automatic slideshow
        cv.waitKey(1000) 

# Close all OpenCV windows
cv.destroyAllWindows()
```
